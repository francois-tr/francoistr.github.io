The goal of this study is to find some relationships and correlations to help a bank manager.
This bank manager has issues since more and more of his customers are leaving their credit card services.
He wants to find a pattern to find if some of his current customer will or will not leave so he can react proactively with better services.

The data set is 10 000 rows and 23 columns with a lot of different informations : age, salary, marital status, etc...

Let's dive into it ! 

FIRST STEP:
Cleaning the data

```python
#After eyeballing my csv file in sublime text, I can see the delimiter is ',' and there is no comments, only a header.

import pandas as pd
pd.set_option('display.max_columns',None)

with open('BankChurners.csv','r') as data:
    Data_Bank = pd.read_csv(data, delimiter = ',')
    Data_Customers = Data_Bank.copy()

Data_Customers
```

```python
#Let's check doublons :
Data_Customers['CLIENTNUM'].nunique()
```

```python
#The numbers of unique values in CLIENTNUM matches the numbers of rows, therefore we do not have any doublons.
#Let's check for missing values now.

for column in Data_Customers.columns:
    Is_null_check = Data_Customers[column].isnull().sum()
    if Is_null_check > 0:
        print(f'{column} : Has {Is_null_check} missing values.')
    

Is_null_check
```

```python
#No missing values, no doublons. Let's starts formating.
Data_Customers.dtypes
```

```python
#In case the bank managers wants to use machine learning in the future, let's convert all we can in number.
#since attrition_Flag means 'the customer left or not' we are going to use dummies to say 0 = the customer still here, 1 = customer left, we
#also see a lot of columns where get dummies will be useful, so first, let's create a function :

def get_dummies_column(DataFrame_column):
    return pd.get_dummies(Data_Customers, columns = [DataFrame_column.name], dtype = int, prefix = None)

Data_Customers = get_dummies_column(Data_Customers['Attrition_Flag'])

Data_Customers
```

Not too much for all of them. so let's use our new function and make everything clearer for the future analysis.

```python
Data_Customers = get_dummies_column(Data_Customers['Gender'])
Data_Customers = get_dummies_column(Data_Customers['Dependent_count'])
Data_Customers = get_dummies_column(Data_Customers['Card_Category'])
Data_Customers = get_dummies_column(Data_Customers['Marital_Status'])
Data_Customers = get_dummies_column(Data_Customers['Education_Level'])
Data_Customers = get_dummies_column(Data_Customers['Income_Category'])
Data_Customers
```

Since unknown tell us nothing on the data, let's drop columns with unknow in their labels.

```python
Data_Customers = Data_Customers.drop(columns = [col for col in Data_Customers.columns if 'Unknown' in col])
Data_Customers
```

Now that we have only numbers in our DataFrame, let's rename the columns to have more visibility

```python
def name_fix(current_name,new_name):
    for column in Data_Customers.columns:
        if column == current_name:
            Data_Customers.rename(columns= {current_name:new_name}, inplace = True)
```

```python
name_fix('Gender_F','Female')
Data_Customers
```

```python
name_fix('Gender_M','Male')
name_fix('Attrition_Flag_Attrited Customer','Attrited_Customer')
name_fix('Attrition_Flag_Existing Customer','Existing_Customer')
Data_Customers
```
